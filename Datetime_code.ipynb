```python
import pandas as pd
```

```python
# Read everything as object
df1 = pd.read_csv('mongodb_comments.csv')

# Convert timestamp to datetime
df1['timestamp'] = pd.to_datetime(df1['timestamp'], format='%m/%d/%Y %H:%M')

# Select only 'id' and 'timestamp' columns
df1 = df1[['_id', 'timestamp']]

# Display the first few rows
df1.head()
```

```python
df2 = pd.read_csv('sentiment.csv')
df2.head(5)
```

```python
# Kết hợp 2 DataFrame dựa trên cột 'id'
merged_df = pd.merge(df1, df2, on='_id')
```
